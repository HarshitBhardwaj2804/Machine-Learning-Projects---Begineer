## A simple machine learning project, leverage's the logistic regresion algorithm to classifies mines and rocks using sonar data.

```python
## Importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
## Importing  Sonar Dataset
df = pd.read_csv("sonar.csv", header = None)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
## spliting the dataset in training and testing
X = df.drop(columns = 60, axis = 1)
y = df[60]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Model Creation

```python
# Using SVM
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred_1 = svc.predict(X_test)
```

```python
## Getting accuracy
from sklearn.metrics import accuracy_score, classification_report
score_1 = accuracy_score(y_test, y_pred_1)
report_1 = classification_report(y_test, y_pred_1)
print(score_1*100)
print(report_1)
```

```python
## checking for multicolinearrity
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                col_name = corr_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr

correlation(X_train, 0.90)
```

```python
X_train_updated = X_train.drop([15, 17, 20], axis = 1)
X_test_updated = X_test.drop([15, 17, 20], axis = 1)
```

```python
svc_1 = SVC()
svc_1.fit(X_train_updated, y_train)
y_pred_2 = svc_1.predict(X_test_updated)
```

```python
## Getting accuracy
from sklearn.metrics import accuracy_score, classification_report
score_2 = accuracy_score(y_test, y_pred_2)
report_2 = classification_report(y_test, y_pred_2)
print(score_2*100)
print(report_2)
```

## Observations:
1. We get a accuracy of 82.53% when we use the svc model before multicolinearrity
2. We get a accuracy of 84.12% after applying multicolinearity

## HyperParameter tunning using GridSearchCV

```python
param_dict = {
    "C" : [1, 10, 100],
    "kernel" : ['linear', 'poly', 'rbf', 'sigmoid']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(svc_1, param_grid = param_dict, scoring = 'accuracy', cv = 5)
```

```python
grid.fit(X_train_updated, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_3 = grid.predict(X_test_updated)
```

```python
## Getting accuracy
from sklearn.metrics import accuracy_score, classification_report
score_3 = accuracy_score(y_test, y_pred_3)
report_3 = classification_report(y_test, y_pred_3)
print(score_3*100)
print(report_3)
```

With hyperparameter tunning we get a accuracy of 90.47%
